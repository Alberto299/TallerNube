# 🌐 Cliente Básico — Prueba tu Servidor en la Nube

Este notebook es el **cliente**: se ejecuta en Google Colab (en los servidores de Google) y le habla, a través de Internet, al **servidor** que desplegaste en tu propia VM (GCP o AWS).

Repasa el diagrama *'El hardware detrás de la pantalla'* de la presentación: aquí Colab hace el papel de **Cliente**.

**Antes de empezar:** reemplaza `IP_SERVIDOR` con la IP pública de tu VM y asegúrate de que el servidor esté corriendo (`uvicorn app:app --host 0.0.0.0 --port 8000`).

```
# Configura aquí la IP pública de tu servidor
IP_SERVIDOR = "TU-IP-PUBLICA"  # ej: "34.123.45.67"
PUERTO = 8000
BASE_URL = f"http://{IP_SERVIDOR}:{PUERTO}"
print("Conectando a:", BASE_URL)
```

## 1. ¿Hay alguien ahí? — GET `/`

```
import requests

respuesta = requests.get(f"{BASE_URL}/")
print("Código de estado:", respuesta.status_code)
print("Respuesta JSON:", respuesta.json())
```

## 2. Chequeo de salud — GET `/health`

```
respuesta = requests.get(f"{BASE_URL}/health")
respuesta.json()
```

## 3. El 'mesero' (API) — POST `/echo`

Enviamos un mensaje en formato JSON. El servidor lo recibe, lo traduce/procesa y regresa una respuesta — igual que el mesero de la diapositiva *'Traductores y memoria del sistema'*.

```
payload = {"mensaje": "Hola servidor, soy un cliente desde Colab"}

respuesta = requests.post(f"{BASE_URL}/echo", json=payload)
print("Código de estado:", respuesta.status_code)
respuesta.json()
```
