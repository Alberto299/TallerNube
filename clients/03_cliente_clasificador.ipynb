# 🤖 Cliente de Clasificación — Simulando un Modelo en la Nube

Este notebook envía un arreglo de números al servidor, que simula un modelo de Machine Learning y regresa una clasificación — el mismo flujo del proyecto *Diagnóstico Cardíaco Asistido por IA* (Fase 3: Procesamiento con VM y API).

```
IP_SERVIDOR = "TU-IP-PUBLICA"
PUERTO = 8000
BASE_URL = f"http://{IP_SERVIDOR}:{PUERTO}"

import requests
print("Conectando a:", BASE_URL)
```

## 1. Generar datos de prueba (simulando una señal)

```
import numpy as np

# Simula, por ejemplo, una pequeña señal de audio o PPG
señal = np.random.normal(loc=0.5, scale=0.15, size=50).tolist()
señal[:5]
```

## 2. Enviar la señal al servidor — POST `/clasificar`

```
payload = {"valores": señal}

respuesta = requests.post(f"{BASE_URL}/clasificar", json=payload)
print("Código de estado:", respuesta.status_code)
resultado = respuesta.json()
resultado
```

## 3. Visualizar la señal enviada y el resultado

```
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.plot(señal)
plt.axhline(resultado["promedio"], color="orange", linestyle="--",
            label=f'Promedio: {resultado["promedio"]}')
plt.title(f'Clasificación del servidor: {resultado["clasificacion"]}')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
